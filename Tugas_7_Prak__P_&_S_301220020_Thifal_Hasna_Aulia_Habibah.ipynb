1. Studi Kasus

Dalam suatu ekosistem, terdapat populasi kelinci sebagai mangsa dan populasi serigala sebagai predator. Populasi kedua spesies ini saling memengaruhi: kelinci sebagai mangsa memiliki kecenderungan untuk bertambah dengan cepat, tetapi pertumbuhannya juga dikendalikan oleh jumlah serigala. Di sisi lain, serigala membutuhkan kelinci sebagai sumber makanan untuk bertahan hidup dan berkembang biak, tetapi populasinya akan menurun jika populasi kelinci berkurang drastis.

**Tujuan:** Mengembangkan model untuk mempelajari dinamika populasi kedua spesies ini dan menentukan kondisi-kondisi yang menyebabkan stabilitas atau ketidakstabilan dalam ekosistem.

3. Penyelesaian

Kita akan menggunakan model Lotka-Volterra dengan persamaan diferensial berikut untuk memodelkan interaksi antara kelinci dan serigala.

Model Lotka-Volterra

Misalkan:


* x adalah jumlah populasi kelinci
 (mangsa).


*  y adalah jumlah populasi serigala (predator).


* α adalah laju pertumbuhan kelinci tanpa adanya serigala.


* β adalah laju di mana serigala menangkap kelinci (laju predasi).


* δ adalah laju di mana serigala bertambah dari kelinci yang ditangkap.


* γ adalah laju kematian serigala tanpa adanya kelinci.

Implementasi dalam Python

Berikut adalah kode untuk mensimulasikan dinamika populasi kelinci dan serigala menggunakan model Lotka-Volterra dengan scipy.integrate.odeint.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameter model
alpha = 0.1   # Laju pertumbuhan kelinci
beta = 0.02   # Laju predasi oleh serigala
delta = 0.01  # Laju reproduksi serigala dari kelinci yang dimakan
gamma = 0.1   # Laju kematian serigala

# Fungsi untuk persamaan diferensial
def lotka_volterra(X, t, alpha, beta, delta, gamma):
    x, y = X
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y
    return [dxdt, dydt]

# Kondisi awal populasi kelinci dan serigala
x0 = 40  # populasi kelinci awal
y0 = 9   # populasi serigala awal
initial_conditions = [x0, y0]

# Waktu simulasi
t = np.linspace(0, 200, 1000)

# Penyelesaian persamaan diferensial
sol = odeint(lotka_volterra, initial_conditions, t, args=(alpha, beta, delta, gamma))

# 1. Pola pertumbuhan populasi kelinci dan serigala dari waktu ke waktu
plt.figure(figsize=(12, 5))
plt.plot(t, sol[:, 0], label="Populasi Kelinci")
plt.plot(t, sol[:, 1], label="Populasi Serigala")
plt.xlabel("Waktu")
plt.ylabel("Populasi")
plt.title("Simulasi Model Lotka-Volterra: Populasi Kelinci dan Serigala")
plt.legend()
plt.grid()
plt.show()

# Menambahkan analisis pola pertumbuhan
if sol[-1, 0] > sol[0, 0] and sol[-1, 1] > sol[0, 1]:
    print("Populasi kelinci dan serigala meningkat sepanjang waktu.")
else:
    print("Populasi kelinci dan serigala menunjukkan pola siklik.")
```
